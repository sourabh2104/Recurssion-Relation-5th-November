```python
# 1.Find the value of T(2) for the recorrence relation T(n) = 3T(n-1) + 12n, given that T(0)=5.


def recurrence_relation(n):
    if n == 0:
        return 5  # Base case: T(0) = 5
    else:
        return 3 * recurrence_relation(n - 1) + 12 * n

# Find the value of T(2)
result = recurrence_relation(2)
print("T(2) =", result)
```

Explanation:-

This code defines a recursive function recurrence_relation that calculates the value of T(n) based on the given recurrence relation.
The base case is T(0)=5, and for any other n, it uses the recurrence relation T3(Tn(1)n) = -+12n.
The code then calculates and prints the value of T(2).



# 2. Given a rec%rrence relation, solve it %sing the s%bstit%tion method:
''' a. T(n) = T(n-1) + c
    b. T(n) = 2T(n/2) + n
    c. T(n) = 2T(n/2) + c
    d. T(n) = T(n/2) + c '''

a. T(n) = T(n-1) + c

Assumption: T(0) is a known constant.

Solution:
T(n) = T(n-1) + c
     = [T(n-2) + c] + c
     = T(n-2) + 2c
     = [T(n-3) + c] + 2c
     = T(n-3) + 3c
     = ...
     = T(0) + nc
     = T(0) + c * n

So, T(n) = T(0) + c * n.

b. T(n) = 2T(n/2) + n

Assumption: T(1) is a known constant.

Solution:
T(n) = 2T(n/2) + n
     = 2[2T(n/4) + n/2] + n
     = 4T(n/4) + 2n
     = 2^k T(n/2^k) + kn

Let k = log_2 n, then 2^k = n. Substituting:
T(n) = 2^log_2 n * T(1) + n * log_2 n
     = n * T(1) + n * log_2 n

So, T(n) = n * T(1) + n * log_2 n.

c. T(n) = 2T(n/2) + c

Assumption: T(1) is a known constant.

Solution:
T(n) = 2T(n/2) + c
     = 2[2T(n/4) + c] + c
     = 4T(n/4) + 2c
     = 2^k T(n/2^k) + kc

Using k = log_2 n, then 2^k = n:
T(n) = 2^log_2 n * T(1) + n * c
     = n * T(1) + nc

So, T(n) = n * T(1) + nc.

d. T(n) = T(n/2) + c

Assumption: T(1) is a known constant.

Solution:
T(n) = T(n/2) + c
     = [T(n/4) + c] + c
     = T(n/4) + 2c
     = [T(n/8) + c] + 2c
     = T(n/8) + 3c
     = ...
     = T(1) + log_2 n * c

So, T(n) = T(1) + log_2 n * c.



```python
# 2. Given a rec%rrence relation, solve it %sing the s%bstit%tion method:
#a. T(n) = T(n-1) + c
#b. T(n) = 2T(n/2) + n
#c. T(n) = 2T(n/2) + c
#d. T(n) = T(n/2) + c 


# Substitution method for solving recurrence relations

def recurrence_a(n, c, T0):
    """
    Solve the recurrence relation T(n) = T(n-1) + c
    with the initial condition T(0) = T0.
    """
    if n == 0:
        return T0
    else:
        return recurrence_a(n-1, c, T0) + c

def recurrence_b(n, T1):
    """
    Solve the recurrence relation T(n) = 2T(n/2) + n
    with the initial condition T(1) = T1.
    """
    if n == 1:
        return T1
    else:
        return 2 * recurrence_b(n//2, T1) + n

def recurrence_c(n, c, T1):
    """
    Solve the recurrence relation T(n) = 2T(n/2) + c
    with the initial condition T(1) = T1.
    """
    if n == 1:
        return T1
    else:
        return 2 * recurrence_c(n//2, c, T1) + c

def recurrence_d(n, c, T1):
    """
    Solve the recurrence relation T(n) = T(n/2) + c
    with the initial condition T(1) = T1.
    """
    if n == 1:
        return T1
    else:
        return recurrence_d(n//2, c, T1) + c

# Test the solutions

# Given values
c_value = 3
T0_value = 5
T1_value = 1

# Test recurrence_a
result_a = recurrence_a(2, c_value, T0_value)
print(f"Recurrence a: T(2) = {result_a}")

# Test recurrence_b
result_b = recurrence_b(4, T1_value)
print(f"Recurrence b: T(4) = {result_b}")

# Test recurrence_c
result_c = recurrence_c(4, c_value, T1_value)
print(f"Recurrence c: T(4) = {result_c}")

# Test recurrence_d
result_d = recurrence_d(8, c_value, T1_value)
print(f"Recurrence d: T(8) = {result_d}")



''' Replace the values of c_value, T0_value, and T1_value with the appropriate constants for your specific recurrence relations. The code uses recursive functions to solve each recurrence relation based on the substitution method.'''
```

```python
# 3. Given a rec%rrence relation, solve it %sing the rec%rsive tree approach:
 # a. T(n) = 2T(n-1) +1
 # b. T(n) = 2T(n/2) + n use python for this


import matplotlib.pyplot as plt
import networkx as nx

def recurrence_a(n):
    """
    Solve the recurrence relation T(n) = 2T(n-1) + 1.
    """
    if n == 0:
        return 1
    else:
        return 2 * recurrence_a(n-1) + 1

def draw_recursive_tree_a(G, parent, n, depth):
    """
    Draw the recursive tree for the recurrence relation T(n) = 2T(n-1) + 1.
    """
    if depth == 0:
        return

    child1 = parent * 2
    child2 = parent * 2 + 1

    G.add_edge(parent, child1, label='2')
    G.add_edge(parent, child2, label='1')

    draw_recursive_tree_a(G, child1, n-1, depth-1)
    draw_recursive_tree_a(G, child2, n-1, depth-1)

def recurrence_b(n):
    """
    Solve the recurrence relation T(n) = 2T(n/2) + n.
    """
    if n == 1:
        return 1
    else:
        return 2 * recurrence_b(n//2) + n

def draw_recursive_tree_b(G, parent, n, depth):
    """
    Draw the recursive tree for the recurrence relation T(n) = 2T(n/2) + n.
    """
    if depth == 0:
        return

    child1 = parent * 2
    child2 = parent * 2 + 1

    G.add_edge(parent, child1, label='2')
    G.add_edge(parent, child2, label=f'n/{2**(depth-1)}')

    draw_recursive_tree_b(G, child1, n//2, depth-1)
    draw_recursive_tree_b(G, child2, n//2, depth-1)

# Test the solutions and draw recursive trees

# Test recurrence_a
result_a = recurrence_a(3)
print(f"Recurrence a: T(3) = {result_a}")

# Draw recursive tree for recurrence_a
G_a = nx.DiGraph()
G_a.add_node(1)
draw_recursive_tree_a(G_a, 1, 3, 3)
pos_a = nx.spring_layout(G_a)
labels_a = nx.get_edge_attributes(G_a, 'label')
nx.draw_networkx(G_a, pos_a, with_labels=True, node_size=700, node_color='skyblue', font_size=10)
nx.draw_networkx_edge_labels(G_a, pos_a, edge_labels=labels_a)
plt.title('Recursive Tree for Recurrence a')
plt.show()

# Test recurrence_b
result_b = recurrence_b(4)
print(f"Recurrence b: T(4) = {result_b}")

# Draw recursive tree for recurrence_b
G_b = nx.DiGraph()
G_b.add_node(1)
draw_recursive_tree_b(G_b, 1, 4, 3)
pos_b = nx.spring_layout(G_b)
labels_b = nx.get_edge_attributes(G_b, 'label')
nx.draw_networkx(G_b, pos_b, with_labels=True, node_size=700, node_color='skyblue', font_size=10)
nx.draw_networkx_edge_labels(G_b, pos_b, edge_labels=labels_b)
plt.title('Recursive Tree for Recurrence b')
plt.show()




''' This code uses the matplotlib and networkx libraries to draw recursive trees for the given recurrence relations. The recursive trees visually represent the recursive structure of the relations. Adjust the input values as needed for your specific recurrence relations.'''
```
